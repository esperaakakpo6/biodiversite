**Fusion des données**

```python
biodiv_grand_est = pd.concat([oiseaux, papillons, plantes], ignore_index=True)
```

```python
print(f"{biodiv_grand_est.info()} \nNombre de doublons: {biodiv_grand_est.duplicated().sum()}")
```

```python
# Conversion
biodiv_grand_est["dateObservation"] = pd.to_datetime(biodiv_grand_est["dateObservation"])

for i in ["cdNom", "codeInseeCommune", "obsID", "idSINPOccTax"]:
    biodiv_grand_est[i] = biodiv_grand_est[i].astype(object)
    
for i in ["latitude", "longitude"]:
    biodiv_grand_est[i] = biodiv_grand_est[i].str.replace(',','.').astype(float)
```

**Merger avec commune.csv**

```python
# communes = pd.read_csv("communes.csv")
communes = pd.read_csv(r"D:\Dataviz\02_Donnees_Secondaires\communes.csv", sep=";")
communes = communes.rename(columns={"codeINSEE" : "codeInseeCommune"})
 
for i in ["latCentroide", "lonCentroide", "urbain", "agricole", "naturel", "humide", "eau", "surfaceTotale"]:
    communes[i] = (
        communes[i]
        .astype(str)
        .str.strip()
        .str.replace(',', '.', regex=False)
    )
    communes[i] = pd.to_numeric(communes[i], errors='coerce')

  
print(f"{communes.info()},\n")
communes.head(2)
```

```python
# colonne communes
colonnes_communes = "codeInseeCommune"
# colonnes à merger
colonnes_merger = ["population 2025", "urbain", "agricole", "naturel", "humide", "eau"]
# merger
biodiv_merger = pd.merge(biodiv_grand_est, communes[[colonnes_communes] + colonnes_merger], on = colonnes_communes, how='left')
```

Statistiques Globales
Objectif : Nombre d'observations par étiquette, espèce, famille, genre ; distributions temporelle et spatiale.

1.1 Nombre d'observations par étiquette, espèce, famille, genre

```python
# Assurez-vous que df est chargé
df = pd.read_csv('biodiv_grand_est_merger.csv', parse_dates=['dateObservation'])

# Statistiques globales
print("Nombre total d'observations :", len(df))

# Par étiquette (oiseau, papillon, etc. - adaptez si d'autres étiquettes existent)
stats_etiquette = df['etiquette'].value_counts()
print("\nObservations par étiquette :\n", stats_etiquette)

# Par espèce
stats_espece = df['espece'].value_counts().head(10)  # Top 10 pour résumé
print("\nTop 10 espèces observées :\n", stats_espece)

# Par famille
stats_famille = df['famille'].value_counts().head(10)
print("\nTop 10 familles :\n", stats_famille)

# Par genre
stats_genre = df['genre'].value_counts().head(10)
print("\nTop 10 genres :\n", stats_genre)

# Graphique : Histogramme des top espèces
stats_espece.plot(kind='bar', figsize=(12, 4))
plt.title("Top 10 Espèces Observées")
plt.xlabel("Espèce")
plt.ylabel("Nombre d'observations")
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

1.2 Distribution temporelle (par année, mois, saison)

```python
df['mois'] = df['dateObservation'].dt.month
df['annee'] = df['dateObservation'].dt.year

# Ajouter colonne saison
def get_saison(mois):
    if mois in [3, 4, 5]:
        return 'Printemps'
    elif mois in [6, 7, 8]:
        return 'Été'
    elif mois in [9, 10, 11]:
        return 'Automne'
    else:
        return 'Hiver'

df['saison'] = df['mois'].apply(get_saison)

# Par année
stats_annee = df['annee'].value_counts().sort_index()
print("\nObservations par année :\n", stats_annee)

# Par mois
stats_mois = df['mois'].value_counts().sort_index()
print("\nObservations par mois :\n", stats_mois)

# Par saison
stats_saison = df['saison'].value_counts()
print("\nObservations par saison :\n", stats_saison)

# Graphiques
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
stats_annee.plot(kind='bar', ax=axes[0], title="Observations par Année")
stats_mois.plot(kind='bar', ax=axes[1], title="Observations par Mois")
stats_saison.plot(kind='bar', ax=axes[2], title="Observations par Saison")
plt.tight_layout()
plt.show()
```

1.3 Distribution spatiale (par département/commune, avec densité)

```python
# Par département
stats_dept = df['departement'].value_counts().head(10)
print("\nTop 10 départements par observations :\n", stats_dept)

# Par commune
stats_commune = df['commune'].value_counts().head(10)
print("\nTop 10 communes par observations :\n", stats_commune)

# Densité : Observations / population 2025 (grouper par commune)
df_densite = df.groupby('commune').agg({
    'population 2025': 'mean',  # Moyenne si multiple
    'obsID': 'count'  # Nombre d'observations
}).rename(columns={'obsID': 'observations'})
df_densite['densite'] = df_densite['observations'] / df_densite['population 2025']
df_densite = df_densite.sort_values('densite', ascending=False).head(10)
print("\nTop 10 communes par densité (obs/population) :\n", df_densite)

# Graphique : Boxplot des densités
plt.figure(figsize=(10, 5))
df_densite['densite'].plot(kind='box')
plt.title("Boxplot des Densités Spatiales (Top 10 Communes)")
plt.ylabel("Densité (obs/population)")
plt.show()
```

3. Prise en Compte des Biais

Objectif : Normaliser par effort d'observation et détecter outliers.

3.1 Effort d'observation et normalisation

```python
# Effort : Nombre unique d'obsID par commune/année
effort = df.groupby(['commune', 'annee'])['obsID'].nunique().reset_index(name='effort')

# Merger avec observations totales
obs_total = df.groupby(['commune', 'annee']).size().reset_index(name='observations')
df_biais = pd.merge(obs_total, effort, on=['commune', 'annee'])
df_biais['normalise'] = df_biais['observations'] / df_biais['effort']
print("\nObservations normalisées par effort :\n", df_biais.sort_values('normalise', ascending=False).head(10))

# Graphique
plt.figure(figsize=(12, 3))
df_biais['normalise'].hist(bins=20)
plt.title("Histogramme des Observations Normalisées")
plt.xlabel("Observations / Effort")
plt.show()
```

3.2 Détection de surplus/outliers en densité

```python
# Outliers via IQR (Interquartile Range)
Q1 = df_densite['densite'].quantile(0.25)  # De 1.3
Q3 = df_densite['densite'].quantile(0.75)
IQR = Q3 - Q1
outliers = df_densite[(df_densite['densite'] < (Q1 - 1.5 * IQR)) | (df_densite['densite'] > (Q3 + 1.5 * IQR))]
print("\nCommunes avec pics anormaux (outliers en densité) :\n", outliers)

# Graphique : Boxplot pour outliers
sns.boxplot(y=df_densite['densite'])
plt.title("Détection d'Outliers en Densité Spatiale")
plt.show()
```
